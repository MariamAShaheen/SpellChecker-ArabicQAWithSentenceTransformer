```python
## test with one pararaph
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F
import re
#Mean Pooling - Take average of all tokens
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output.last_hidden_state #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

#Encode text
def encode(texts):
    # Tokenize sentences
    encoded_input = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')

    # Compute token embeddings
    with torch.no_grad():
        model_output = model(**encoded_input, return_dict=True)

    # Perform pooling
    embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

    # Normalize embeddings
    embeddings = F.normalize(embeddings, p=2, dim=1)

    return embeddings


# Sentences we want sentence embeddings for
# query = "How many people live in London?"

query = "التصنيع ده بيستخدم ايه بشكل كتير اوي؟"
paragraph = """
صناعة الورق على مستوى العالم، ده موضوع كبير ومهم جداً. الورق من الحاجات الأساسية في حياتنا اليومية، من الكتب للجرايد للمستندات وغيرها. الورق بدأ من الصين قديماً، وبعدها انتشر في كل العالم.
في مصر، الورق ليه تاريخ طويل، خصوصاً مع ورق البردي اللي كان بيستخدم قديماً. النهاردة، صناعة الورق في مصر بتستخدم تقنيات حديثة وبتنتج أنواع كتير من الورق.
على مستوى العالم، صناعة الورق بتواجه تحديات كبيرة، خصوصاً مع الاهتمام بالحفاظ على البيئة. صناعة الورق بتستهلك كميات كبيرة من الأشجار، وده بيسبب مشاكل بيئية زي قطع الأشجار وتلوث المياه.
"""

docs = re.split(r'[،]\s*', paragraph.strip())
# Load model from HuggingFace Hub
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/multi-qa-mpnet-base-cos-v1")
model = AutoModel.from_pretrained("sentence-transformers/multi-qa-mpnet-base-cos-v1")

#Encode query and docs
query_emb = encode(query)
doc_emb = encode(docs)

#Compute dot score between query and all document embeddings
scores = torch.mm(query_emb, doc_emb.transpose(0, 1))[0].cpu().tolist()

#Combine docs & scores
doc_score_pairs = list(zip(docs, scores))

#Sort by decreasing score
doc_score_pairs = sorted(doc_score_pairs, key=lambda x: x[1], reverse=True)

#Output passages & scores
for doc, score in doc_score_pairs:
    print(score, doc)
import csv

# Define output file path
output_file = "/content/sample_data/output.csv"

# Write the output to the CSV file
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Question", "Answer", "Score"])  # Write header row

    for doc, score in doc_score_pairs:
        writer.writerow([query, doc, score])
```

```python
##use model to predict the first ranked 10 answer for each question
## multilingual-e5-small perform well
# output_files with different models ["all-MiniLM-L6-v2-output.csv", "all-MiniLM-L6-v2-output.csv", "distilbert-base-multilingual-cased-output.csv", "multilingual-e5-small.csv", "multi-qa-MiniLM-L6-cos-v1-output.csv", "multi-qa-mpnet-base-cos-v1-output"]
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F
import re
import csv
#Mean Pooling - Take average of all tokens
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output.last_hidden_state #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

#Encode text
def encode(texts):
    # Tokenize sentences
    encoded_input = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')

    # Compute token embeddings
    with torch.no_grad():
        model_output = model(**encoded_input, return_dict=True)

    # Perform pooling
    embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

    # Normalize embeddings
    embeddings = F.normalize(embeddings, p=2, dim=1)

    return embeddings


tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/multi-qa-MiniLM-L6-cos-v1")
model = AutoModel.from_pretrained("sentence-transformers/multi-qa-MiniLM-L6-cos-v1")
training_data_file = "/content/drive/MyDrive/widebot/Similarity"
output_file = "/content/drive/MyDrive/widebot/multi-qa-MiniLM-L6-cos-v1-output.csv"
with open(output_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["doc_id","Question", "Answer", "Score"])  # Write header row
    with open(training_data_file, mode='r', newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Skip header row
        doc_id= 0
        for row in reader:
          doc_id +=1
          print("doc_id", doc_id)
          paragraph = row[0]
          questions = row[1:]
          docs = re.split(r'[،]\s*', paragraph.strip())  # try adding , to splitting but not goof=d
          for query in questions:
              #Encode query and docs
              query_emb = encode(query)
              doc_emb = encode(docs)

              #Compute dot score between query and all document embeddings
              scores = torch.mm(query_emb, doc_emb.transpose(0, 1))[0].cpu().tolist()

              #Combine docs & scores
              doc_score_pairs = list(zip(docs, scores))

              #Sort by decreasing score
              doc_score_pairs = sorted(doc_score_pairs, key=lambda x: x[1], reverse=True)
              for doc, score in doc_score_pairs[:10]:
                  writer.writerow([doc_id,query, doc, score])
```

```python
#After using transform based model to try to use a ground truth based on frequency of answer among all models this file is created instead of test_data
import pandas as pd
base_path = '/content/drive/MyDrive/widebot/'
# List of CSV file paths
csv_files = ["all-MiniLM-L6-v2-output.csv",  "distilbert-base-multilingual-cased-output.csv", "multilingual-e5-small.csv",
             "multi-qa-MiniLM-L6-cos-v1-output.csv", "multi-qa-mpnet-base-cos-v1-output.csv", "AraElectra-Arabic-SQuADv2-QA-output.csv"]  # Add your CSV file paths here

# Read all CSV files into a list of dataframes
dfs = [pd.read_csv(base_path+file) for file in csv_files]

# Concatenate all dataframes into a single dataframe
df_concatenated = pd.concat(dfs, ignore_index=True)

# Group by document ID and question, and aggregate the answers with frequency
grouped_df = df_concatenated.groupby(['doc_id', 'Question', 'Answer']).size().reset_index(name='frequency')
sorted_df = grouped_df.sort_values(by='frequency', ascending=False)
# Write the grouped dataframe to a new CSV file
output_file = "ranked_answers.csv"
sorted_df.to_csv(output_file, index=False)
```
